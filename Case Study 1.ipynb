## Importing data

```python
import pandas as pd
import numpy as np
```

```python
path = 'C:/Users/onkar/OneDrive - somaiya.edu/Desktop/area - Sheet1.csv'
```

```python
df = pd.read_csv(path)
```

## Spliting required column into multiple columns

```python
df1 = df['area'].str.split(',', expand=True)
```

```python
df1
```

```python
header = ['a','b','c','d','e','f','g','h','i','j','k','l']
```

```python
df1.columns = header
df1.columns
```

```python
df1
```

## Combining room and area of room

```python
df1['room1-Area(sq.ft)'] = df1['a'] + '-' + df1['b']
```

```python
df1['room2-Area(sq.ft)'] = df1['c']+'-'+df1['d']
```

```python
df1['room3-Area(sq.ft)'] = df1['e']+'-'+df1['f']
```

```python
df1['room4-Area(sq.ft)'] = df1['g']+'-'+df1['h']
```

```python
df1['room5-Area(sq.ft)'] = df1['i']+'-'+df1['j']
```

```python
df1['room6-Area(sq.ft)'] = df1['k']+'-'+df1['l']
```

```python
cd  = df1.drop(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l'], axis=1)
```

```python
cd.replace(np.nan, 0, inplace=True)
```

```python
cd
```

## Concatenating splitted columns with Names 

```python
dff = pd.concat([df,cd], axis=1)
dff
```

```python
dff.drop(['area'], axis=1, inplace=True)
```

```python
dff
```

```python
dff.to_csv('Clean_dff.csv')
```
