```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('D:/Coding Files/Data Science/Projects/LGM-December-Data-Science-Internship/sp.jpg')
```

```python
plt.imshow(img)
plt.axis(False)
plt.show()
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

### Converting to gray image

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

### Inveting the gray image

```python
invert_img = cv2.bitwise_not(gray_img)
```

### Bluring the inverting image

```python
blur_img = cv2.GaussianBlur(invert_img, (111,111),0)
```

### Inverting blur image

```python
invblur_img = cv2.bitwise_not(blur_img)
```

### Sketch image

```python
sketch_img = cv2.divide(gray_img, invblur_img, scale=256.0)
```

```python
cv2.imwrite('sk.png', sketch_img)
```

```python
cv2.imshow('sk image', sketch_img)
cv2.waitKey(0)
```

```python
cv2.destroyAllWindows()
```

```python
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')
```

```python
plt.title('Sketch image', size=18)
rgb_sk = cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sk)
plt.axis('off')
```
